```python
import tkinter as tk 


# Create new instance window of tkinter Tk class
window = tk.Tk()

# Set window title 
window.title("Test GUI")

# Create widgets: Frames
frame_top = tk.Frame(master=window, relief="sunken", borderwidth=2)
frame_left = tk.Frame(master=window, relief="sunken", borderwidth=2)


# Create a widget: Label 
label = tk.Label(
    text="Hello, Tkinter",
    fg="black",
    bg="lightblue",
    master=frame_top
)

# Create a widget: Entry
entry = tk.Entry(fg="blue", bg="lightblue", width=50, master=frame_left)


# Create a widget: Button 
button = tk.Button(
    text="Click me!",
    width=25,
    height=5,
    bg="lightblue",
    fg="black",
    master=frame_left
)

# Create a widget: Text
text_box = tk.Text() 


# Adding widgets to window and configure parameter
frame_top.pack(side=tk.TOP, anchor=tk.W, fill=tk.X, expand=tk.YES)
frame_left.pack(side=tk.LEFT)
label.pack()
entry.pack()
button.pack()
text_box.pack()


# Method to listen for events 
window.mainloop()


```
