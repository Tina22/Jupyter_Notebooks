# Matplotlib Diagramms 

```python
# Imports
import matplotlib.pyplot as plt
import numpy as np

# global custom setting for the running commands matplotlib
from matplotlib import rcParams
plt.rcParams['axes.titlepad'] = 50 
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
```

## Function Graphs - Big O Notation

```python
fig = plt.figure()  

plt.title("Big O-Notation")
fig.suptitle('Shows some Bachmannâ€“Landau notation graphically')  

x = np.linspace(0.01,10, 100)

plt.hlines(y=5, xmin=0, xmax=100, label='constant')
plt.plot(x, np.log(x), label='log naturalis')
plt.plot(x, np.log2(x), label='log basis 2')
plt.plot(x, x, label='linear')
plt.plot(x, np.power(x,2), label='quadratic')
plt.plot(x, np.power(x,3), label='qubic')
plt.plot(x, np.exp(x), label='exponential')

plt.axis([0,10,0,10])

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.savefig("Big_O_Notation.png")
plt.show()
```

## Numpy Mathematical Functions

### Sinus

```python
x = np.arange(0, 15, 0.1)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)

plt.savefig("Sinus_Graph.png")

plt.show()
```

### Cosinus

```python
x = np.arange(0, 15, 0.1)
y = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y)

plt.savefig("Cosinus_Graph.png")

plt.show()
```

### Multiple graphs in one plot

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.sin(2 * np.pi * x1) * np.exp(-x1)
y2 = np.sin(2 * np.pi * x2)

#plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-', label='Damped oscillation')
plt.plot(x2, y2, '.-', label='Undamped oscillation')
plt.xlabel('time (s)')

plt.legend()
plt.savefig("Damped_Undamped_Oscillation.png")
plt.show()
```

### Histogramm

```python
gaussian_numbers = np.random.normal(loc=3.0,scale=1.0,size=1000)
plt.hist(gaussian_numbers)
plt.title("Gaussian Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.savefig("Gaussian_Histogram.png")
plt.show()
```
