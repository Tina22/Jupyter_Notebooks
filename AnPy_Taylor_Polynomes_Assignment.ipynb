# Taylor Polynome 

```python
import math
import sympy as sy
import numpy as np
from sympy.utilities.lambdify import lambdify
from sympy.functions import sin,cos
import matplotlib.pyplot as plt
```

```python
class function(sy.Function):
    @classmethod
    def eval(cls, x):
        return sin(x)
```

```python
plt.style.use("ggplot")
x = sy.Symbol('x')

# Taylor Approximation für x0 für vordefinierte Funktion 
def taylor(f, x0, n):
    p = 0
    for i in range(1, n + 1):
        df = f(x).diff(x, i)
        p += df.subs(x, x0) / math.factorial(i) * (x - x0)**i
    return p
    
```

```python
def plot_one(x1,f,label):
    y = []
    for k in x1:
        # Für alle x aus linspace, die y Koordinate berechnen
        y.append(f.subs(x,k))
    # Plotten der Funktion
    plt.plot(x1,y,label=label)
```

```python
def plot():
    x1 = np.linspace(-5,5,500)
   # Plotten der Original Funktion
    plot_one(x1, function(x), "Original Funktion")
        
    # Annäherung bis 10, beginnend bei 1, in zweier Schritten
    for j in range(1,10,2):
        result = taylor(function,0,j)
        print('Taylor Approximation bei n='+str(j),result)
        
        # Plotten der Annäherungen 
        plot_one(x1, result, 'Grad '+str(j))
       
        
    plt.xlim([-5,5])
    plt.ylim([-5,5])
    plt.xlabel('x')
    plt.ylabel('y')
    
    plt.legend()
    plt.grid(True)
    plt.title('Taylorreihen Approximation')
    
    plt.show()
```

```python
#main Funktion  
def main():
        
    plot()

if __name__ == "__main__":
    main()
```
